```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#loading iris dataset
iris = datasets.load_iris()
data = iris.data
```

```python
#k means
kmeans = KMeans(n_clusters=3, init='k-means++',max_iter=300,n_init=10,random_state=0)
y_means = kmeans.fit_predict(data)
```

```python
d=fig, ax = plt.subplots(figsize = (15,10))
plt.scatter(data[y_means == 0,0], data[y_means == 0,1], s=50, c='pink', label='Setosa')
plt.scatter(data[y_means == 1,0], data[y_means == 1,1], s=50, c='green', label='Versicolor')
plt.scatter(data[y_means == 2,0], data[y_means == 2,1], s=50, c='blue', label='Virginica')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='orange', label ='Centroids', edgecolors='k')
plt.legend()
```
